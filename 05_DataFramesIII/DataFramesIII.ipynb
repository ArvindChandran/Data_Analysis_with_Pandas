### 5. The Catch-All .ix[] Method

```python
bond = pd.read_csv("../00_Datasers/jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
bond.ix["GoldenEye"]
```

```python
bond.ix[["Diamonds are Forever", "Moonraker", "Spectre"]]
```

```python
bond.ix["A View to a Kill" : "The World Is Not Enough"]
```

```python
# bond.ix["Secred Bond"]
bond.ix[["Spectre", "Secred Bond"]]
```

```python
"Spectre" in bond.index
```

```python
"Secred Bond" in bond.index
```

```python
bond.ix[10:15]
bond.ix[[8,16,24]]
# bond.ix[100]
# bond.ix[[8, 30]]
```

### 6. Second Arguments to .loc[], .iloc[] and .ix[] Methods

```python
bond = pd.read_csv("../00_Datasers/jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
bond.loc["Moonraker", "Director" : "Budget"]
```

```python
bond.loc["Moonraker", ["Actor", "Budget", "Year"]]
```

```python
bond.iloc[14, 2 : 5]
```

```python
bond.iloc[14, [5, 3, 2]]
```

```python
bond.ix[20, "Budget"]
```

```python
bond.ix["The Man with the Golden Gun", 2]
```
