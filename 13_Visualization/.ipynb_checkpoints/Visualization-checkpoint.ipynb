# Visualization

```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. The .plot() Method

```python
bb = data.DataReader(name = "BBRY", data_source="yahoo", start="1995-05-01", end = "2016-12-31")
bb.head(3)
```

```python
bb[["High", "Low"]].plot(figsize=(14,8))
#bb.plot(y = "High", figsize=(14,8))
```

### 2. Modifying Aesthetics with Templates

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
bb.plot(y="Close", figsize=(14,8))
```

```python
plt.style.use('dark_background')
bb.plot(y="Close", figsize=(14,8))
```

```python
plt.style.use('ggplot')
bb.plot(y="Close", figsize=(14,8))
```

### 3. Bar Graphs

```python
google = data.DataReader(name = "GOOG", data_source="yahoo", start="2004-01-01", end = "2017-12-31")
google.head(3)
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 and stock_price <= 500:
        return "Satisfactory"
    else:
        return "Steller"
```

```python
google["Close"].apply(rank_performance).value_counts().plot(kind = "barh")
```

### 4. Pie Charts

```python
apple = data.DataReader(name = "AAPL", data_source="yahoo", start="2012-01-01", end = "2017-12-31")
apple.head(3)
```

```python
mean = apple["Close"].mean()
```

```python
def rank_performence(stock_price):
    if stock_price <= mean:
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use("ggplot")
apple["Close"].apply(rank_performence).value_counts().plot(kind = "pie", legend = True)
```

### 5. Histograms

```python
google = data.DataReader(name = "GOOG", data_source="yahoo", start="2004-01-01", end = "2017-12-31")
google.head(3)
```

```python
def custom_round(stock_price):
    return int(stock_price / 100.00) * 100
```

```python
google["High"].apply(custom_round).value_counts().sort_index()
```

```python
google["High"].apply(custom_round).nunique()
```

```python
google["High"].apply(custom_round).plot(kind="hist", bins = 10)
```
